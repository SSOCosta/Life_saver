# IMAGE LABELING

```python
# Defining a function that labels all the images inside of each folder with the folder name and a distcint number
def rename_images(folder_path, prefix):
    # Get the list of files in the folder
    files = os.listdir(folder_path)

    # Sort files to ensure consistent numbering
    files.sort()

    # Iterate through each file in the folder
    for index, file_name in enumerate(files, start=1):
        # Generate the new file name
        new_name = f"{prefix}_{index}{os.path.splitext(file_name)[1]}"

        # Construct the full paths
        old_path = os.path.join(folder_path, file_name)
        new_path = os.path.join(folder_path, new_name)

        # Check if the destination file already exists
        while os.path.exists(new_path):
            index += 1
            new_name = f"{prefix}_{index}{os.path.splitext(file_name)[1]}"
            new_path = os.path.join(folder_path, new_name)

        # Print the renaming operation
        print(f"Renaming {old_path} to {new_path}")

        # Rename the file
        os.rename(old_path, new_path)

def rename_all_images(root_folder):
    for category_folder in os.listdir(root_folder):
        category_path = os.path.join(root_folder, category_folder)

        if os.path.isdir(category_path):
            print(f"\nProcessing images in folder: {category_path}")
            rename_images(category_path, prefix=category_folder)

# Specify the root folder
root_folder = 'C:/Users/susy_/Desktop/Final Project/image_galery'

# Call the function to rename images in all folders
rename_all_images(root_folder)
```

# SPLITTING DATA ( X/y and Train Test)

```python
import os
import shutil
import random
from sklearn.model_selection import train_test_split

ALLOWED_EXTENSIONS = {'.png', '.jpg', '.jpeg', '.gif', '.bmp'}

def has_allowed_extension(file):
    _, extension = os.path.splitext(file)
    return extension.lower() in ALLOWED_EXTENSIONS

def split_dataset(root_folder, train_size=0.7, validation_size=0.2, random_state=42):
    X = []  # X (Features)
    y = []  # Y (Labels)
    train_folder = os.path.join(root_folder, 'train')
    valid_folder = os.path.join(root_folder, 'valid')
    test_folder = os.path.join(root_folder, 'test')

    # Create destination folders if they don't exist
    for folder in [train_folder, valid_folder, test_folder]:
        os.makedirs(folder, exist_ok=True)

    for category_folder in os.listdir(root_folder):
        category_path = os.path.join(root_folder, category_folder)
        if category_folder in ['train', 'valid', 'test']:
            continue

        if os.path.isdir(category_path):
            # Create subfolders within train, valid, and test folders
            train_category_folder = os.path.join(train_folder, category_folder)
            valid_category_folder = os.path.join(valid_folder, category_folder)
            test_category_folder = os.path.join(test_folder, category_folder)

            for subfolder in [train_category_folder, valid_category_folder, test_category_folder]:
                os.makedirs(subfolder, exist_ok=True)

            image_files = [file for file in os.listdir(category_path) if has_allowed_extension(file)]

            # Shuffle the image files
            random.shuffle(image_files)

            # Check if there are enough samples for splitting
            if len(image_files) < 2:
                print(f"   Skipping category '{category_folder}' due to insufficient samples.")
                continue

            # Split image files into training, validation, and test sets: Train-test-split
            train_files, valid_test_files = train_test_split(image_files, train_size=train_size, random_state=random_state)
            valid_files, test_files = train_test_split(valid_test_files, test_size=0.1, random_state=random_state)

            print(f"Category '{category_folder}':")
            print(f" - Total images: {len(image_files)}")
            print(f" - Train images: {len(train_files)}")
            print(f" - Valid images: {len(valid_files)}")
            print(f" - Test images: {len(test_files)}")

            # Move files to respective subfolders
            for file in train_files:
                src_path = os.path.join(category_path, file)
                dest_path = os.path.join(train_category_folder, file)
                shutil.move(src_path, dest_path)

            for file in valid_files:
                src_path = os.path.join(category_path, file)
                dest_path = os.path.join(valid_category_folder, file)
                shutil.move(src_path, dest_path)

            for file in test_files:
                src_path = os.path.join(category_path, file)
                dest_path = os.path.join(test_category_folder, file)
                shutil.move(src_path, dest_path)

if __name__ == "__main__":
    root_folder = "image_galery"
    split_dataset(root_folder)
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

model = Sequential()  # Create a Sequential model
def build_moderate_model():
    model.add(Conv2D(16, (3, 3), 1, activation='relu', input_shape=(256, 256, 3)))
    model.add(MaxPooling2D())
    model.add(Conv2D(32, (3, 3), 1, activation='relu'))
    model.add(MaxPooling2D())
    model.add(Conv2D(16, (3, 3), 1, activation='relu'))
    model.add(MaxPooling2D())
    model.add(Flatten())
    model.add(Dense(256, activation='relu'))
    model.add(Dense(10, activation='softmax'))  # Use 'softmax' for multi-class classification

    # Compile with adam optimizer & categorical_crossentropy loss function
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Build and summarize the model
moderate_model = build_moderate_model()
print(moderate_model.summary())
```

```python
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras import layers
from tensorflow.keras.models import Model

def build_inception_model():
    # Load InceptionV3 model with pre-trained weights (excluding the top layer)
    base_model = InceptionV3(weights='imagenet', include_top=False, input_shape=(256, 256, 3))

    # Freeze the pre-trained layers
    for layer in base_model.layers:
        layer.trainable = False

    # Create a new model on top
    x = layers.Flatten()(base_model.output)
    x = layers.Dense(256, activation='relu')(x)
    output = layers.Dense(10, activation='softmax')(x)  # Use 'softmax' for multi-class classification

    # Combine the base model with the new layers
    model = Model(inputs=base_model.input, outputs=output)

    # Compile the model with adam optimizer & categorical_crossentropy loss function
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    return model

# Build and summarize the Inception model
inception_model = build_inception_model()
print(inception_model.summary())
```
