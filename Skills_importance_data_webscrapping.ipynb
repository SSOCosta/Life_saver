'''The European Skills Index (ESI) is Cedefop’s composite indicator measuring the performance of EU skills systems. 
The ESI measures countries’ “distance to the ideal” performance. This ideal performance is chosen as the highest
achieved by any country over a period of 7 years. The ideal performance is scaled to be 100 and the scores of all 
countries are then computed and compared to that. The ESI consists of three pillars; skills development, activation 
and matching, each of which measures a different aspect of a skills system. Basis of the ESI are 15 individual indicators
from various international datasets. The scores are calculated across countries at the indicators’ level. The scores are 
then averaged at the various layers and finally the Index score is formed. To illustrate, an Index (or pillar, sub-pillar etc.) 
score of 65 suggests that the country has reached 65% of the ideal performance. Thus, there is still 35% (100-65) room for 
improvement. A score of 100 corresponds to achieving the ‘frontier’, that is an aspirational target performance for that 
indicator. A score of 0 corresponds to a lowest-case performance. The shaded part in the graphs below shows how far (in %) 
a country has scored in comparison to the ideal, while the dotted line shows what is the remaining distance to cover until 
the best performance is reached.'''

# WEBSCRAPPING

```python
#importing the necessary libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# using page url
url = "https://keydifferences.com/difference-between-hard-skills-and-soft-skills.html"
```

```python
# download html with a get request
response = requests.get(url)
response.status_code # 200 status code means OK!
```

```python
# parse html (create the 'soup')
soup = BeautifulSoup(response.content, "html.parser")
```

```python
soup
```

```python
#print(soup.prettify())
```

```python
soup.select("body > div")
```

```python
print(soup.select("body > div > div.site-inner > div > main > article > div > h3:nth-child(9)")[0].get_text())
```

```python
print(soup.select("body > div > div.site-inner > div > main > article > div > p:nth-child(1)")[0].get_text())
```

```python
print(soup.select("#tablepress-1015")[0].get_text())
```

```python
#Transforming text in the proper dataframe
pd.set_option('display.max_colwidth', None)
data = []
# Extract the table with id 'tablepress-1015'
table = soup.select("#tablepress-1015")[0]
# Extract rows from the table
rows = table.find_all('tr')
# Extract data from each row
for row in rows[1:]:  # Skip the header row
    columns = row.find_all(['th', 'td'])
    data.append([col.get_text(strip=True) for col in columns])
# Create a DataFrame from the extracted data
df = pd.DataFrame(data, columns=['Basis for Comparison', 'Hard Skills', 'Soft Skills'])
# Print the DataFrame
df
```

```python
#importing the necessary libraries
import requests
from bs4 import BeautifulSoup
from io import BytesIO
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# to get image from Soft skills vs Hard skills
url = "https://resumekit.com/blog/hard-skills-vs-soft-skills/"
```

```python
# download html with a get request
response = requests.get(url)
print(response)
```

```python
# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Find the image tag
img_tag = soup.find('img', {'src': 'https://resumekit.com/blog/wp-content/uploads/2022/04/hard-soft-skills-examples.png'})
# Get the image source URL
img_url = img_tag['src']
```

```python
# Download the image
img_response = requests.get(img_url)
img_data = BytesIO(img_response.content)
img = Image.open(img_data)
# # Open and display the image using PIL (Python Imaging Library)
# img.show()
```

```python
# Plot the image using Matplotlib
plt.imshow(img)
plt.axis('off')  # Turn off axis labels
plt.show()
```
